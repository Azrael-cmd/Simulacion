```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
aleatorios = np.random.uniform(0, 1, 100)
ks_statistic, p_value = stats.kstest(aleatorios, 'uniform')
print("Números aleatorios generados:")
print(aleatorios)
# resultados
print(f'Estadística de Kolmogorov-Smirnov: {ks_statistic}')
print(f'Valor p: {p_value}')
# interpretacion
alpha = 0.05  # nivel de significancia
if p_value > alpha:
    print("los datos parecen venir de una distribución uniforme.")
else:
    print("os datos no siguen una distribución uniforme.")
plt.hist(aleatorios, bins=10, edgecolor='black', alpha=0.7)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Distribución de Números Aleatorios')
plt.show()
```
