###Contexto
Una estación de carga rápida para vehículos eléctricos quiere
simular el tiempo entre la llegada de los vehículos para prever la cantidad
de puntos de carga necesarios durante el día. La llegada de los vehículos
sigue una distribución exponencial con una tasa de 1 llegada cada 15
minutos, en promedio.

### Objetivo
Generar tiempos entre llegadas de vehículos eléctricos durante
un día de operación (12 horas).

```python
import numpy as np
import matplotlib.pyplot as plt
lambda_llegadas = 1 / 15
num_horas = 12
num_minutos = num_horas * 60
U = np.random.uniform(0, 1, num_minutos)
tiempos_entre_llegadas = -np.log(1 - U) / lambda_llegadas
print("Tiempos entre llegadas de vehículos simulados (en minutos):")
print(tiempos_entre_llegadas)
plt.hist(tiempos_entre_llegadas, bins=30, edgecolor='black')
plt.title("Distribución de Tiempos entre Llegadas de Vehículos Eléctricos (Distribución Exponencial)")
plt.xlabel("Tiempo entre llegadas (minutos)")
plt.ylabel("Frecuencia")
plt.show()
```
