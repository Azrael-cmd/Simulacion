```python
import numpy as np
import matplotlib.pyplot as plt
lambda_servicio = 1 / 2
def generar_tiempo_servicio(lambda_servicio):
    u = np.random.uniform(0, 1)
    tiempo_servicio = -1 / lambda_servicio * np.log(1 - u)
    return tiempo_servicio
tiempos_servicio = [generar_tiempo_servicio(lambda_servicio) for _ in range(100)]
print("Tiempos de servicio generados:", tiempos_servicio)
plt.hist(tiempos_servicio, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos de Servicio Simulados (Exponencial)')
plt.xlabel('Tiempo de servicio (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
lambda_servicio = 1 / 3
def generar_tiempo_servicio(lambda_servicio):
    u = np.random.uniform(0, 1)
    tiempo_servicio = -1 / lambda_servicio * np.log(1 - u)
    return tiempo_servicio
tiempos_servicio = [generar_tiempo_servicio(lambda_servicio) for _ in range(30)]
print("Tiempos de servicio generados:", tiempos_servicio)
plt.hist(tiempos_servicio, bins=5, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos de Servicio Simulados (Exponencial)')
plt.xlabel('Tiempo de servicio (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```

### Beneficio de uso de variables aleatorias
El uso de variables aleatorias en la simulación de tiempos de servicio permite modelar escenarios realistas en sistemas de colas. La distribución exponencial es ideal para representar tiempos de atención variables, facilitando el análisis del rendimiento del sistema, la optimización de recursos y la evaluación de distintos escenarios sin pruebas en entornos reales. En el código, se generan tiempos de servicio aleatorios con
λ=1/2, obteniendo una distribución típica donde la mayoría de los tiempos son cortos, con algunos valores más largos.
