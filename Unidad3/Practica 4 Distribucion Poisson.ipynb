```python
import numpy as np
import matplotlib.pyplot as plt
import math
lambda_poisson = 5
n = 100
U = np.random.uniform(0, 1, n)
def poisson_inverse_transform(lambda_poisson, U):
    k = 0
    F_k = np.exp(-lambda_poisson)
    while U > F_k:
        k += 1
        F_k += (lambda_poisson**k * np.exp(-lambda_poisson)) / math.factorial(k)
    return k
demanda_simulada = [poisson_inverse_transform(lambda_poisson, u) for u in U]
print("Demanda diaria simulada:", demanda_simulada)
plt.hist(demanda_simulada, bins=np.arange(0, max(demanda_simulada) + 1) - 0.5, edgecolor='black', density=True)
plt.title("Simulación de demanda diaria (Distribución Poisson)")
plt.xlabel("Número de productos demandados")
plt.ylabel("Frecuencia")
plt.show()
```

### Beneficio de uso de variables aleatorias
El uso de variables aleatorias en la distribución de Poisson permite modelar eventos discretos en intervalos de tiempo o espacio, como la demanda de productos o la llegada de clientes. Facilita simulaciones realistas, optimiza la planificación de recursos y se aplica en diversas áreas como logística y telecomunicaciones. Generar valores con λ=5 usando la transformación inversa, muestra en un histograma la distribución esperada. Esto ayuda a tomar decisiones basadas en datos simulados antes de aplicarlas en sistemas reales.
